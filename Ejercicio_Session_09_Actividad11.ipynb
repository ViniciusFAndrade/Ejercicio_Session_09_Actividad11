```
#Substituí os inputs por números aleatórios e reduzi o range de possibilidades para que o programa não rodasse por muito tempo

import os
import random

#Variáveis
import_item = 0
subtotal = 0
total = 0
desconto = 0
cont_compra = 0
outra_c = "S"

#Gerar Ticket
os.system("clear")

outra_c = "S"
while outra_c == "S":
    while True:
        cont_compra += 1
        import_item = random.randint(0,10)
        if import_item == 0:
            cont_compra -= 1
            break
        elif import_item < 0:
            cont_compra -= 1
            print("Número negativo")
            continue
        else:
            subtotal += import_item
        
        #Cálculo Desconto
        if subtotal > 10:
            desconto = subtotal * 0.1
        total = subtotal - desconto

        #Mostrar totais
        os.system("clear")
        print("Detalhes da Compra")
        print(" ")
        print("Quantidade: ", cont_compra)
        print("Subtotal: ", subtotal)
        print("Desconto: ", desconto)
        if desconto > 0:
            print("###Desconto aplicado a compras acima de 10###")
        print(" ")
        print("Valor Total: ",total)
        print(" ")

    #Simulando o input de uma nova compra
    if random.randint(0,10) > 4:
        outra_c = "S"
    else:
        outra_c = "N"
    break

os.system("clear")
print("*"*50)
print("Obrigado, volte sempre!")
    
```
